### Comments:

The program works very well for our picture setup with a white paper background.
It is able to detect texts with varying Backgrounds, fonts & Colors. However, the last example demonstrates a mis-detection. The EAST text detector from OpenCV did not detect the bounding box region corretly for the small green number on the house and instead identified the fence, pytesseract recognises the text inside the fence bounding box as a group of letters m & n. 

We can say, overall it performs well, but it can be improved. This application can be used in a variety of applications such as number plate detection, street signs and house numbers to name a few.

Future work will be building our custom OCR from scratch using deep learning. Example would be a Resnet50 and KNN.
